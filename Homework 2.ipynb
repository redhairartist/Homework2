```python
def sum_distance(from_num, to_num):

    if from_num > to_num:
        from_num, to_num = to_num, from_num
    total = sum(range(from_num, to_num + 1))

    return total

print (sum_distance(2,8))

"""Если from_num больше to_num, мы меняем их местами, чтобы всегда суммировать от меньшего к большему. range создает последовательность чисел, sum суммирует их
```

```python
def trim_and_repeat(string, offset=0, repetitions=1):
    trimmed_string = string[offset:]
    result = trimmed_string * repetitions
    return result
print(trim_and_repeat("Svetlana Menchuk", 8, 3))

"""string - это строка, котрую нужно обрезать и повторить. offset = 0 обозначает количество символов, которые нужно обрезать слева (по умолчанию = 0). repetitions = 1 это количество повторений строки (по умолчанию 1). Команда string[offset:] — возвращает часть строки, начиная с символа с индексом offset и до конца строки. trimmed_string * repetitions — повторяет строку trimmed_string указанное количество раз (repetitions)"""
```
